# PAFA reproduction entry point

This notebook is a lightweight, auditable entry point. It reads the pinned paper/source/data receipts before any model action. Large source trees, checkpoints, caches, logs, and predictions remain under the timestamped gitignored result root.


```python
from datetime import datetime
import os
from pathlib import Path
import subprocess
from zoneinfo import ZoneInfo

METHOD = "pafa"
DATASET_ROOT = Path(os.environ.get("ACOUSTIC_DATASET_ROOT", "dataset/raw/icbhi_2017"))
RUN_ROOT = Path(os.environ.get("ACOUSTIC_RUN_ROOT", f"result/{METHOD}_{datetime.now(ZoneInfo('America/Chicago')):%Y%m%d_%H%M%S}"))
DEVICE = os.environ.get("ACOUSTIC_DEVICE", "cuda")
EXECUTE = False  # Set true only after reviewing DATASET_ROOT and RUN_ROOT.

commands = [
    ["python", "-m", "baseline.pafa.run_reproduction", "bootstrap", "--dataset-root", str(DATASET_ROOT), "--result-root", str(RUN_ROOT), "--device", DEVICE],
    ["python", "-m", "baseline.pafa.run_reproduction", "verify-bootstrap", "--result-root", str(RUN_ROOT)],
    ["python", "-m", "baseline.pafa.run_reproduction", "smoke", "--result-root", str(RUN_ROOT), "--device", DEVICE, "--steps", "1"],
    ["python", "-m", "baseline.pafa.run_reproduction", "profile", "--result-root", str(RUN_ROOT), "--device", DEVICE, "--steps", "100"],
    ["python", "-m", "baseline.pafa.run_reproduction", "full", "--result-root", str(RUN_ROOT), "--device", DEVICE],
]
if EXECUTE:
    for command in commands:
        subprocess.run(command, check=True)
commands
```

## Execution policy

Run source/data, environment, preprocessing-forward, profile, and full gates in that order. Do not change the author data, model, loss, preprocessing, or metric semantics. Every compatibility change requires a separate receipt.
